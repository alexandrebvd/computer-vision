```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (10.0, 10.0)
matplotlib.rcParams['image.cmap'] = 'gray'
```

```python
def pencilSketch(image, arguments=0):
    
    #Edge preserving
    image = cv2.edgePreservingFilter(image, flags=1, sigma_s=1.2, sigma_r=0.12)
    
    #Detail enhancement
    image = cv2.detailEnhance(image, sigma_s=5, sigma_r=0.22)
    
    #convert to gray scale
    grayImage = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    #invert the gray image
    grayImageInv = 255 - grayImage

    #Apply gaussian blur
    gaussian = cv2.GaussianBlur(grayImageInv, (47, 47), 0)

    #blend using color dodge
    blend = cv2.divide(grayImage, 255-gaussian, scale=256.0)
    
    # Sharpen kernel
    sharpen = np.array((
        [0, -1, 0],
        [-1, 5, -1],
        [0, -1, 0]), dtype="int")
    
    # Using 2D filter by applying the sharpening kernel
    pencilSketchImage = cv2.filter2D(blend, -1, sharpen)

    return cv2.cvtColor(pencilSketchImage, cv2.COLOR_GRAY2BGR)
```

```python
def cartoonify(image, arguments=0):
    
    #Just use colors from original image with sketch image
    sketch = pencilSketch(image)
    cartoonImage = cv2.bitwise_and(image, sketch)

    return cartoonImage
```

```python
imagePath = "trump.jpg"
image = cv2.imread(imagePath)

cartoonImage = cartoonify(image)
pencilSketchImage = pencilSketch(image)
```

```python
plt.figure(figsize=[20,10])
plt.subplot(131);plt.title("Original image");plt.imshow(image[:,:,::-1]);
plt.subplot(132);plt.title("Pencil Sketch image");plt.imshow(pencilSketchImage[:,:,::-1]);
plt.subplot(133);plt.title("Cartoon image");plt.imshow(cartoonImage[:,:,::-1]);
```
